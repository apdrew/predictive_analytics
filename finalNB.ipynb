# First, deal with the timestamps

```python
def fix_timestamps(pdf):
    nt = []
    for i in pdf.index:
        gutc = pdf.loc[i]['Timestamp'].split('-')[-1].replace(':', '').replace(' Denver', '')
        gds = '-'.join(pdf.loc[i]['Timestamp'].split('-')[0:-1])
        fgs = gds + '-' + gutc
        t = datetime.strptime(fgs, "%Y-%m-%dT%H:%M:%S%z")
        nt.append(t)
    pdf['ts'] = nt    

    return pdf
```

```python
fix_timestamps(pa_hva3_df)
fix_timestamps(pa_elec_df)

print(pa_hva3_df.head())
print(pa_elec_df.head())
```
